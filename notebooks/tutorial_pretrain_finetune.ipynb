# NeuroAugment tutorial
Simulate data, pretrain an encoder, and run a finetune placeholder workflow.

```python
from neuroaugment.simulator import Simulator
from neuroaugment.ssl.trainer import SSLTrainer
sim = Simulator(T=256, C=3, seed=0)
X, y, meta = sim.sample_batch(16)
trainer = SSLTrainer(3, latent_dim=32)
losses = trainer.fit_arrays(X, steps=10, batch_size=4)
losses[-1]
```
